```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
path = 'Questionnaire'
questionnaire_data = pd.read_csv('Questionnaire.csv')
```

```python
#To display the first few rows of the data
print(questionaire_data.head())

#To get a summary of your dataframe
print(questionaire_data.info())
```

```python
print(questionnaire_data.columns)
```

```python
World = questionnaire_data['World'].value_counts()

#Plotting
World.plot(kind='bar', color=['skyblue', 'lightgreen', 'yellow'])  # You can choose different colors
plt.title('Which part of the world have you spent most of your life in?')
plt.xlabel('World')
plt.ylabel('Number of Votes')

#Optional: Add annotations (numbers) on each bar for clarity
for index, value in enumerate(World):
    plt.text(index, value, str(value))

plt.show()
```

```python
Work = questionnaire_data['Work'].value_counts()

#Plotting
Work.plot(kind='bar', color=['skyblue', 'lightgreen', 'yellow', 'magenta', 'purple', 'orange', 'navy'  ])  # You can choose different colors
plt.title('What motivates you to work?')
plt.xlabel('Work')
plt.ylabel('Number of Votes')

plt.show()
```
